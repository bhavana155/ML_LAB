```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the Dataset
df = pd.read_csv("titanic.csv")   # Change filename if needed
```

```python
# Step 3: Display First and Last Rows
df.head()
df.tail()
# Step 4: Check Shape of Dataset
df.shape
# Step 5: Display Column Names
df.columns
# Step 6: Information About Dataset
df.info()
# Step 7: Statistical Summary
df.describe()
# Step 8: Check Missing Values
df.isnull().sum()
# Step 9: Check Duplicate Rows
df.duplicated().sum()
# Step 10: Check Data Types
df.dtypes
# Step 11: Survival Count
df["Survived"].value_counts()
# Visualization
sns.countplot(x="Survived", data=df)
plt.show()
# Step 12: Gender Distribution
sns.countplot(x="Sex", data=df)
plt.show()
# Step 13: Passenger Class Distribution
sns.countplot(x="Pclass", data=df)
plt.show()
# Step 14: Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)
plt.show()
# Step 15: Correlation Matrix
numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
# Step 16: Boxplot for Age
sns.boxplot(x=df["Age"])
plt.show()
# Step 17: Relationship Between Age and Survival
sns.boxplot(x="Survived", y="Age", data=df)
plt.show()
# Step 18: Survival by Gender
sns.countplot(x="Sex", hue="Survived", data=df)
plt.show()
# Step 19: Survival by Passenger Class
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.show()
# Step 20: Fare Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Fare"], bins=20, kde=True)
plt.show()
```
