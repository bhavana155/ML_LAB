```python
#Cell 1: Load Dataset and Inspect
import pandas as pd
import numpy as np

# Load Titanic dataset
titanic_df = pd.read_csv("titanic.csv")

print("Dataset Shape (rows, cols):", titanic_df.shape)

print("\nFirst 3 rows of raw dataset:")
print(titanic_df.head(3))

print("\nMissing Values:")
print(titanic_df.isnull().sum())

print("\nDuplicate Rows:", titanic_df.duplicated().sum())
```

```python
#Cell 2: Handle Missing Values
from sklearn.impute import SimpleImputer

# Fill missing Age values using median
age_imputer = SimpleImputer(strategy='median')
titanic_df[['Age']] = age_imputer.fit_transform(titanic_df[['Age']])

# Fill missing Embarked values using most frequent value
embarked_imputer = SimpleImputer(strategy='most_frequent')
titanic_df[['Embarked']] = embarked_imputer.fit_transform(titanic_df[['Embarked']])

# Fill missing Fare values using median (if any)
fare_imputer = SimpleImputer(strategy='median')
titanic_df[['Fare']] = fare_imputer.fit_transform(titanic_df[['Fare']])

# Drop Cabin column because it contains many missing values
titanic_df.drop(columns=['Cabin'], inplace=True)

print("Missing Values after preprocessing:\n")
print(titanic_df.isnull().sum())
```

```python
#Cell 3: Encode Categorical Features
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

# Encode categorical columns
titanic_df['Sex'] = encoder.fit_transform(titanic_df['Sex'])
titanic_df['Embarked'] = encoder.fit_transform(titanic_df['Embarked'])

print("Encoded Dataset (First 5 Rows):")
print(titanic_df[['Sex', 'Embarked']].head())
```

```python
#Cell 4: Normalize and Standardize Numerical Features
from sklearn.preprocessing import MinMaxScaler, StandardScaler

numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']

# Normalization
minmax_scaler = MinMaxScaler()
normalized_data = minmax_scaler.fit_transform(titanic_df[numerical_features])

print("Normalized Range:", normalized_data.min(), "to", normalized_data.max())

# Standardization
standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(titanic_df[numerical_features])

print("Standardized Mean:", round(standardized_data.mean(),4))
print("Standardized Std Dev:", round(standardized_data.std(),4))
```

```python
#Cell 5: Transform Skewed Feature
from sklearn.preprocessing import PowerTransformer

# Check skewness of Fare
original_skew = titanic_df['Fare'].skew()
print("Original skewness of 'Fare':", round(original_skew,4))

# Apply Yeo-Johnson transformation
power_transformer = PowerTransformer(method='yeo-johnson')
fare_transformed = power_transformer.fit_transform(titanic_df[['Fare']])

transformed_skew = pd.Series(fare_transformed.flatten()).skew()
print("Skewness after Power Transformation:", round(transformed_skew,4))
```

```python
#Cell 6: Remove Duplicate Records
before = titanic_df.shape[0]

titanic_df.drop_duplicates(inplace=True)

after = titanic_df.shape[0]

print("Rows before removing duplicates:", before)
print("Rows after removing duplicates :", after)
```

```python
#Cell 7: Export Preprocessed Dataset
# Create processed DataFrame using standardized numerical features
processed_titanic_df = titanic_df.copy()

processed_titanic_df[numerical_features] = standardized_data

# Save dataset
processed_titanic_df.to_csv("processed_titanic.csv", index=False)

print("Processed Titanic dataset exported successfully to 'processed_titanic.csv'.")
print("Processed Dataset Shape:", processed_titanic_df.shape)
```
